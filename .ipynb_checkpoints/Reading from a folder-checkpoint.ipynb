```python
from PIL import Image 
import pytesseract as pt 
import os 
      
def main(): 
    # path for the folder for getting the raw images 
    path =r"C:\Users\majidmm\Documents\MTI and RC\RC\INPUT FILES\Images\\"
  
    # link to the file in which output needs to be kept 
    fullTempPath =r"C:\Users\majidmm\Documents\MTI and RC\RC\INPUT FILES\textfiles\outputFile.txt"
  
    # iterating the images inside the folder 
    for imageName in os.listdir(path): 
        inputPath = os.path.join(path, imageName) 
        img = cv2.imread(inputPath) 
  
        # applying ocr using pytesseract for python 
        text = pt.image_to_string(img, lang ="eng",config=r' --psm 6') 
  
        # saving the  text for appending it to the output.txt file 
        # a + parameter used for creating the file if not present 
        # and if present then append the text content 
        file1 = open(fullTempPath, "a+") 
  
        # providing the name of the image 
        file1.write(imageName+"\n") 
  
        # providing the content in the image 
        file1.write(text+"\n") 
        file1.close()  
  
    # for printing the output file 
    file2 = open(fullTempPath, 'r') 
    print(file2.read()) 
    file2.close()         
  
  
if __name__ == '__main__': 
    main() 
```

```python
from PIL import Image
import pytesseract
import os
import pandas as pd
import cv2

f = []
t = []
x = []
y = []
z = []

input_dir = r'C:\Users\majidmm\Documents\MTI and RC\RC\INPUT FILES\Images\\'

for root, dirs, filenames in os.walk(input_dir):
    for filename in filenames:
        try:
            print(filename)
            f.append(filename)
            im = cv2.imread(input_dir + filename)
            text = pytesseract.image_to_string(im,config=r' --psm 6')
            t.append(text)
            license_number = text[80:98]
            rental_time = text[758:790]
            return_time = text[827:859]
#             print(text)
            print(license_number)
            print(rental_time)
            print(return_time)
            print('-='*20)
            x.append(license_number)
            y.append(rental_time)
            z.append(return_time)
        except:
            continue


df = pd.DataFrame(list(zip(f, t, x, y, z)),columns=['file_Name','Text','License_Number','Rental_Time','Return_Time'])
df.to_csv(r"C:\Users\majidmm\Documents\MTI and RC\RC\INPUT FILES\textfiles\magma.csv")
```
